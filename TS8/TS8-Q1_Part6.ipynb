```python
import sympy as sym
```

```python
w0, E, I, L, x = sym.symbols('w_0, E, I, L, x')
```

```python
w = w0*x/L
```

```python
v = sym.Function('v')
```

```python
eq = sym.Eq( sym.diff(v(x),(x,4)), -w/(E*I))
display(eq)
```

```python
bc1 = {v(0): 0}

bc2 = {sym.diff(v(x),(x,2)).subs(x,0): 0}

bc3 = {v(L): 0}

bc4 = {sym.diff(v(x),(x,2)).subs(x,L): 0}

bc_list = {**bc1, **bc2, **bc3, **bc4}
display(bc_list)
```

```python
diff_eq_soln = sym.dsolve(eq, v(x), ics=bc_list)
display(diff_eq_soln.simplify())
```

## Alternative solution

```python
R1 = (w0*L/2*(L/3))/L
R2 = w0*L/2 - R1
```

```python
M = R1*x - w*x/2*(x/3)
```

```python
eq = sym.Eq(sym.diff(v(x),(x,2)), M/(E*I))
display(eq)
```

```python
bc_list = {**bc1, **bc3}
```

```python
diff_eq_soln = sym.dsolve(eq, v(x), ics=bc_list)
display(diff_eq_soln.simplify())
```
