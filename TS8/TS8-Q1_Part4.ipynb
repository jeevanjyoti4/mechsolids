```python
import sympy as sym
```

```python
w, E, I, L, x = sym.symbols('w, E, I, L, x')
```

```python
v = sym.Function('v')
```

```python
eq = sym.Eq(sym.diff(v(x),(x,4)), -w/(E*I))
display(eq)
```

```python
bc1 = {v(0): 0}

bc2 = {sym.diff(v(x),(x,2)).subs(x,0): 0}

bc3 = {v(L): 0}

bc4 = {sym.diff(v(x),(x,2)).subs(x,L): 0}

bc_list = {**bc1, **bc2, **bc3, **bc4}
display(bc_list)
```

```python
diff_eq_soln = sym.dsolve(eq, v(x), ics=bc_list)
display(diff_eq_soln.simplify())
```

```python
v_soln = diff_eq_soln.simplify().rhs
display(v_soln)
```

## Alternative solution

```python
R1 = w*L/2
```

```python
M = R1*x-w*x**2/2
```

```python
eq = sym.Eq(sym.diff(v(x),(x,2)),M/(E*I))
display(eq)
```

```python
bc_list = {**bc1, **bc3}
display(bc_list)
```

```python
diff_eq_soln = sym.dsolve(eq,v(x), ics=bc_list)
display(diff_eq_soln.simplify())
```
