```python
import sympy as sym
```

```python
M0, E, I, L, x = sym.symbols('M_0, E, I, L, x')
```

```python
M = M0
```

```python
v = sym.Function('v')
```

```python
eq = sym.Eq( sym.diff(v(x),(x,2)), M/(E*I))
display(eq)
```

```python
bc1 = {v(0): 0}
bc2 = {sym.diff(v(x),x).subs(x,0): 0}

bc_list = {**bc1, **bc2}
display(bc_list)
```

```python
diff_eq_soln = sym.dsolve(eq, v(x), ics=bc_list)
display(diff_eq_soln)
```

```python
v_soln = diff_eq_soln.rhs
display(v_soln)
```
