```python
import sympy as sym
```

```python
M0, E, I, L, x = sym.symbols('M_0, E, I, L, x')
```

```python
R1 = M0/L
```

```python
M = R1*x
```

```python
v = sym.Function('v')
```

```python
eq = sym.Eq(sym.diff(v(x),(x,2)), M/(E*I))
display(eq)
```

```python
bc1 = {v(0): 0}
bc2 = {v(L): 0}

bc_list = {**bc1, **bc2}
display(bc_list)
```

```python
diff_eq_soln = sym.dsolve(eq,v(x),ics=bc_list)
display(diff_eq_soln.simplify())
```

```python
v_soln = diff_eq_soln.simplify().rhs
display(v_soln)
```
