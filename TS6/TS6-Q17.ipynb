```python
from math import pi, sqrt
```

```python
d = 12e-3
t = 0.6e-3
p = 3.5e6
F = 1000
T = 30

r = d/2
A = pi*((r+t)**2 - r**2)
J = pi/2*((r+t)**4 - r**4)

sigma_xx = F/A + p*r/(2*t)
sigma_yy = p*r/t
tau_xy = -T*(r+t)/J

sigma_avg = (sigma_xx + sigma_yy)/2
R = sqrt(((sigma_xx - sigma_yy)/2)**2 + tau_xy**2)

sigma_p1 = sigma_avg+R
sigma_p2 = sigma_avg-R

display(sigma_p1/1e6, sigma_p2/1e6)
```
