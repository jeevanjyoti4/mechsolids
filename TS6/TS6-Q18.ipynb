```python
import numpy as np
```

```python
F0 = np.array([-450, 0, 300])
```

```python
F = -F0
display(F)
```

```python
r = np.array([-0.3,0.2,0.3])
```

```python
M0 = np.cross(r,F0)
```

```python
M = -M0
display(M)
```

```python
Ro = 0.03
Ri = 0.02
A = np.pi*(Ro**2 - Ri**2)
I = 1/4*np.pi*(Ro**4 - Ri**4)
J = 2*I
```

```python
sigma_xxA = np.dot(np.cross(M,[0,-Ro,0])/I,[1,0,0]) + np.dot(F,[1,0,0])/A
display(sigma_xxA/1e6)
```

```python
Q = 1/2*np.pi*Ro**2*4*Ro/(3*np.pi) - 1/2*np.pi*Ri**2*4*Ri/(3*np.pi)
display(Q)
```

```python
tau_xzA = np.dot(np.cross(M,[0,-Ro,0])/J,[0,0,1]) + np.dot(F,[0,0,1])*Q/(I*2*(Ro-Ri))
display(tau_xzA/1e6)
```

```python
stressA = np.matrix([[sigma_xxA/1e6, tau_xzA/1e6],[tau_xzA/1e6, 0]])
display(stressA)
```

```python
[eivalA, eivecA] = np.linalg.eig(stressA)
eivalA
```

```python
sigma1A = eivalA[0]
sigma2A = eivalA[1]
display(sigma1A, sigma2A)
```

```python
tau_maxA_inplane = 1/2*abs(sigma1A - sigma2A)
display(tau_maxA_inplane)
```

```python
sigma_xxB = np.dot(np.cross(M,[0,0,-Ro])/I,[1,0,0]) + np.dot(F,[1,0,0])/A
display(sigma_xxB/1e6)
```

```python
tau_xyB = np.dot(np.cross(M,[0,0,-Ro])/J,[0,1,0]) + np.dot(F,[0,1,0])*Q/(I*2*Ro)
display(tau_xyB/1e6)
```

```python
stressB = np.matrix([[sigma_xxB/1e6, tau_xyB/1e6],[tau_xyB/1e6, 0]])
display(stressB)
```

```python
[eivalB, eivecB] = np.linalg.eig(stressB)
eivalB
```

```python
sigma1B = eivalB[0]
sigma2B = eivalB[1]
display(sigma1B, sigma2B)
```

```python
tau_maxB_inplane = 1/2*abs(sigma1B - sigma2B)
display(tau_maxB_inplane)
```
