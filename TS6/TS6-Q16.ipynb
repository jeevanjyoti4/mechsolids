```python
import numpy as np
```

```python
R = 0.025/2
```

```python
I = 1/4*np.pi*R**4
display(I)
```

```python
F0 = np.array([-350, -375, 450])
```

```python
F = -F0
```

```python
r = np.array([-0.2, 0, 0.075])
```

```python
M0 = np.cross(r,F0)
```

```python
M = -M0
display(M)
```

\begin{gather}
\sigma = \frac{(\vec{M}\times \vec{R}) \cdot \hat{n}}{I} + \frac{\vec{F}\cdot \hat{n}}{A}
\end{gather}

```python
sigma_A = np.dot(np.cross(M,[0,R,0])/I,[1,0,0]) + np.dot(F,[1,0,0])/(np.pi*R**2)
display(sigma_A/1e6)
```

```python
sigma_B = np.dot(np.cross(M,[0,0,R])/I,[1,0,0]) + np.dot(F,[1,0,0])/(np.pi*R**2)
display(sigma_B/1e6)
```

```python
Q = 2/3*R**3
J = 1/2*np.pi*R**4
```

\begin{gather}
\tau = \frac{(\vec{M}\times \vec{R})\cdot \hat{s}}{J} + \frac{(\vec{F}\cdot \hat{s}) Q}{It}
\end{gather}

```python
tau_xzA = np.dot(np.cross(M,[0,R,0])/J,[0,0,1]) + np.dot(F,[0,0,1])*Q/(I*2*R)
display(tau_xzA/1e6)
```

```python
tau_xyB = np.dot(np.cross(M,[0,0,R])/J,[0,1,0]) + np.dot(F,[0,1,0])*Q/(I*2*R)
display(tau_xyB/1e6)
```
