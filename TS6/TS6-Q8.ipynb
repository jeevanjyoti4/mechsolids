We consider $x$ along the logitudinal direction and $y$ along the circumferential (or, hoop) direction.

```python
from math import pi, sqrt, sin, cos
import sympy as sym
```

```python
bbeta = 30*pi/180
tt = 8e-3
sigma_all = 75e6
dout = 600e-3
rr = dout/2 - tt
display(tt,rr)
```

\begin{gather}
\sigma_{x^\prime x^\prime} = \sigma_{xx} \cos^2 \theta + \sigma_{yy} \sin^2 \theta + 2\tau_{xy} \sin\theta \cos\theta
\end{gather}

```python
p, r, t, beta = sym.symbols('p, r, t, beta')
```

```python
theta = beta
sigma_xx = p*r/(2*t)
sigma_yy = p*r/(t)
tau_xy = 0
display(sigma_xx, sigma_yy)
```

```python
sigma_xdash = (sigma_xx+sigma_yy)/2 + (sigma_xx - sigma_yy)/2*(sym.cos(2*theta)) + tau_xy*sym.sin(2*theta)
display(sigma_xdash)

sigma_xdash_interim = sigma_xdash.subs([(r,rr),(t,tt),(beta,bbeta)])
display(sigma_xdash_interim)
```

```python
eq = sym.Eq(sigma_xdash_interim, sigma_all)
display(eq)
soln, = sym.solve(eq,[p])

display(soln/1e6)
```
