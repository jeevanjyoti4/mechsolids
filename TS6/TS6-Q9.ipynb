```python
import sympy as sym
from math import pi
```

```python
p, r, t, theta, E, nu = sym.symbols('p, r, t, theta, E, nu')
```

```python
sigma_xx = p*r/(2*t)
sigma_yy = p*r/t
tau_xy = 0
```

```python
eps_xx = 1/E*(sigma_xx - nu*sigma_yy)
eps_yy = 1/E*(sigma_yy - nu*sigma_xx)
eps_xy = 0
display(eps_xx, eps_yy)
```

```python
eps_xxdash = (eps_xx + eps_yy)/2 + (eps_xx - eps_yy)/2*sym.cos(2*theta) + eps_xy*sym.sin(2*theta)
display(eps_xxdash)
```

```python
rr = 1.5/2
tt = 25e-3
EE = 200e9
nnu = 1/3
ttheta = 45*pi/180
eps_a = 100e-6
eps_b = 250e-6
```

```python
eps_xxdash_value = eps_xxdash.subs([(r,rr),(t,tt),(E,EE),(nu,nnu),(theta,ttheta)])
eq = sym.Eq(eps_xxdash_value, eps_b)
sym.solve(eq,[p])[0]/1e6
```
