```python
from math import pi
```

```python
b = 200
tw = 25
tf = 50
h = 200
s = 75
d = 4
V = 2e3

ybar = ( (h-tf/2)*b*tf + (h/2*h*tw)*2 )/(b*tf+h*tw*2)
display(ybar)

I = 1/12*b*tf**3 + b*tf*(h-tf/2-ybar)**2 + (1/12*tw*h**3 + tw*h*(h/2-ybar)**2)*2
display(I)
```

```python
Q1 = (h-tf/2-ybar)*b*tf  #horizontal board
display(Q1)

Q2 = (h/2-ybar)*tw*h*2  # both the vertical boards
display(Q2)

F = s*V*Q1/(2*I)
tau = F/(pi*d**2/4)
display(tau/1e6)
```
