```python
bt = 50
ht = 50

tw = 50
hw = 150

b = 150
tf = 50

A = b*(hw + 2*tf) - 2*bt*hw

I = (1/12)*b*(hw + 2*tf)**3 - 2*(1/12)*bt*hw**3
display(I)
```

## For B:

Consider the lower board.

```python
Q1 = b*tf*(hw/2 + tf - tf/2)

tau_all = 1.5
t1 = tw

V1 = tau_all*I*t1/(Q1)
display(V1)
```

## For D:

Consider the left bit of the upper flange. 

```python
Q2 = bt*tf*(hw+tf/2)

t2 = tf

V2 = tau_all*I*t2/(Q2)
display(V2)
```

Therefore, the maximum allowable shear force $V$ will be $\min(V1,V2) = 16.7$ kN.
