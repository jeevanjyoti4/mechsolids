```python
tw = 40
h = 250
b = 250
tf = 25
ht = 75
tt = 25

F = 500
V = 3.5e3
```

Referred to the top:

Area of the web: $A_1 = h t_w$

Area of the flange: $A_2 = b t_f$

Area of the tips: $A_3 = 2h_t t_t$  (because there are two tips)

Centroid of the web: $y_1 = t_f + \dfrac{h}{2}$

Centroid of the flange: $y_2 = \dfrac{t_f}{2}$

Centroid of the tips: $y_3 = \dfrac{h_t}{2}$

Centroid of the overall cross-section:
$\bar{y} = \dfrac{(A_1 y_1 + A_2 y_2 + A_3 y_3)}{(A_1 + A_2 + A_3)}$

```python
A1 = h*tw
A2 = b*tf
A3 = 2*ht*tt  

y1 = tf+h/2
y2 = tf/2
y3 = ht/2

ybar = (A1*y1 + A2*y2 + A3*y3)/(A1 + A2 + A3)
display(ybar)
```

The neutral axis (NA) will pass through the centroid. 

Distance of the centroid of the web from the neutral axis: $d_1 = |\bar{y} - y_1|$

Distance of the centroid of the flange from the neutral axis: $d_2 = |\bar{y} - y_2|$

Distance of the centroid of the tips from the neutral axis: $d_1 = |\bar{y} - y_3|$

Area moment of inertia of the web: $I_1 = \dfrac{1}{12}t_w h^3 + A_1 d_1^2$

Area moment of inertia of the flange: $I_2 = \dfrac{1}{12}b t_f^3 + A_2 d_2^2$

Area moment of inertia of the tips: $I_1 = 2\times\dfrac{1}{12}t_t h_t^3 + A_3 d_3^2$

```python
d1 = abs(ybar - y1)
d2 = abs(ybar - y2)
d3 = abs(ybar - y3)

I1 = (1/12)*tw*h**3 + A1*d1**2
I2 = (1/12)*b*tf**3 + A2*d2**2
I3 = 2*(1/12)*tt*ht**3 + A3*d3**2
I = I1 + I2 + I3   #Unites in mm^4
display(I)
```

## __For the calculation of $s$:__

We can consider either the shaded portion in the figure below or the unshaded portion. 

Here, we will show the calculations considering the shaded portion first and then the unshaded portion, and demonstrate that the magnitude is the same in both cases. 

__NOTE:__ In the exam, you have to consider only one of these. 




_**Considering the shaded portion**_

Area of the shaded portion: $A_{1s} = t_t h_t$

Note that we are considering only one bit. If we consider both bits, then at this stage we will have to multiply by 2. 

Location of the centroid of the shaded portion referred to the NA of the overall cross-section: $y_{1s} = \bar{y} - y_3$

Therefore, we have:

$Q_{1s} = A_{1s} y_{1s}$

The shear flow will be $\dfrac{V Q_{1s}}{I}$. 

But this shear flow must be equal to the shear force that each nail can support per unit length, i.e. $\dfrac{F}{s}$. 

(Had we considered the area of both bits earlier, then at this stage we would have to consider $2\dfrac{F}{s}$. 

So, we obtain: $\dfrac{F}{s} = \dfrac{V Q_{1s}}{I}$, which implies $s = \dfrac{F I}{V Q_{1s}}$

```python
# For s considering shaded portion:
A1s = tt*ht
y1s = ybar - y3
Q1s = A1s*y1s 
s1 = F*I/(V*Q1s)  #Units of mm
display(s1)
```

_**Considering the unshaded portion**_

The $Q$ for the unshaded portion will be calculated by considering the web and the flange. 

We have: 

$Q_{2s} = A_1 (y_1 - \bar{y}) + A_2 (y_2 - \bar{y})$

This time both rows of nails will have to be considered so that. 

Equating the shear flow to the force supported by both rows of nails per unit length (into the plane of the screen), we have:

$\dfrac{V Q_{2s}}{I} = 2 \dfrac{F}{s}$ 

from which we obtain:

$s = 2 \dfrac{F I}{V Q_{2s}}$

```python
# For s considering unshaded portion:

Q2s = A1*(y1 - ybar) + A2*(y2 - ybar) # Note that Q2s will be negative of Q1s
s2 = 2*F*I/(V*Q2s)  #Units of mm
display(s2)
```

_**Considering the shaded portion**_

The $Q$ for the shaded portion will be calculated by considering the flange and both the bits. 

We have: 

$Q_{1s'} = A_2 (\bar{y} - y_2) + A_3 (\bar{y} - y_3)$

Equating the shear flow to the force (supported by the nails) per unit length (into the plane of the screen), we have:

$\dfrac{V Q_{1s'}}{I} = \dfrac{F}{s'}$ 

from which we obtain:

$s' = \dfrac{F I}{V Q_{1s'}}$

```python
Q1sdash = A2*(ybar - y2) + A3*(ybar - y3)
sdash1 = F*I/(V*Q1sdash)  #Units of mm
display(sdash1)
```

_**Considering the unshaded portion**_

The unshaded portion is basically just the web. 

Area of the unshaded portion: $A_{2s'} = A_1$

Location of the centroid of the unshaded portion referred to the NA of the overall cross-section: $y_{2s'} = y_1 - \bar{y}$

Therefore, we have:

$Q_{2s'} = A_{2s'} y_{2s'}$

The shear flow will be $\dfrac{V Q_{2s'}}{I}$. 

But this shear flow must be equal to the shear force that each nail can support per unit length, i.e. $\dfrac{F}{s'}$. 

So, we obtain: $\dfrac{F}{s'} = \dfrac{V Q_{2s'}}{I}$, which implies $s' = \dfrac{F I}{V Q_{2s'}}$

```python
# For sdash considering unshaded portion:

Q2sdash = A1*(y1 - ybar)  #Note that Q2sdash will be negative of Q1sdash
sdash2 = F*I/(V*Q2sdash)  #Units of mm
display(sdash2)  
```
