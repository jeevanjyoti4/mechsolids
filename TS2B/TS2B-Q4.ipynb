```python
b3 = 25
h3 = 100
A3 = b3*h3
y3 = h1+h2+h3/2
```

```python
A = A1 + A2 + A3 #total area
```

```python
ybar = (A1*y1 + A2*y2 + A3*y3)/A
display(ybar)
```

Distance of each of the individual centroids from the neutral axis:

```python
d1 = abs(ybar - y1)
d2 = abs(ybar - y2)
d3 = abs(ybar - y3)
```

```python
I1 = 1/12*(2*b1*h1**3)+A1*d1**2
I2 = 1/12*b2*h2**3 + A2*d2**2
I3 = 1/12*b3*h3**3 + A3*d3**2

I = I1+I2+I3
display(I)
```

```python
M = 3000e3  # Units in N-m
ytop = h1 + h2 + h3 - ybar
ybottom = -ybar 
sigma_top = -M*ytop/I
sigma_bottom = -M*ybottom/I

display(sigma_top, sigma_bottom)
```
