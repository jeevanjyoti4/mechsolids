```python
from math import pi

P = 150e3
omega = 800*2*pi/60  #rev/min has to be converted to rad/s

R = 0.1/2
L_BC = 3
L_CD = 4
G = 75e9

J = pi*R**4/2


T_BC = 1*P/omega  #torque in BC is due to 100% of P
T_CD = 0.30*P/omega  #torque in CD is due to 30% of P because 70% of P is taken away by the gear at C

tau = T_BC*R/J
display(tau/1e6) #divide by 10^6 to give the value in MPa

phi_CB = T_BC*L_BC/(G*J)
phi_DC = T_CD*L_CD/(G*J)

phi_E = phi_D = phi_DC + phi_CB   #phi_E = phi_D because there is no torque in DE
display(phi_E*180/pi)  #converted from rad to degrees
```
